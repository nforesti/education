```python
%matplotlib inline

# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

# City Datasets

```python
cities_dataset_path = './datasets/cities/'
counties_dataset_path = './datasets/counties/'

# create lists with names of all city and county csv files
city_files = (c for c in os.listdir(cities_dataset_path) if c.endswith(".csv"))
county_files = (c for c in os.listdir(counties_dataset_path) if c.endswith("county.csv"))

# create a dictionary with mapping between city/county name and csv data
city = dict()
for f in city_files:
    city[f] = pd.read_csv(cities_dataset_path + f)
    
county = dict()
for f in county_files:
    county[f] = pd.read_csv(counties_dataset_path + f)

# check
print(city['stanford-ca.csv'])

print(county['alameda-ca-county.csv'])
```
